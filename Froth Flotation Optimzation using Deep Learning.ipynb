RNN Architecture

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM, SimpleRNN
```

```python
from keras.preprocessing.sequence import TimeseriesGenerator
 
n_input = 6
n_features = 2
generator = TimeseriesGenerator(data=scaled_train,
                                targets=scaled_train,
                                length=n_input,
                                batch_size=32)
```

```python
RNN_model=Sequential()
RNN_model.add(SimpleRNN(30, input_shape=(6, 2), activation='relu', return_sequences=True))
RNN_model.add(SimpleRNN(30, activation='relu', return_sequences=True))
RNN_model.add(SimpleRNN(30, activation='relu', return_sequences=False))
RNN_model.add(Dense(2, activation='relu'))
```

```python
RNN_model.summary()
```

```python
RNN_model.compile(loss='mean_squared_error',optimizer='Adam')
```

```python
RNN_model_history=RNN_model.fit_generator(generator,epochs=50)
```

```python
import matplotlib.pyplot as plt




plt.plot(RNN_model_history.history['loss'])
plt.title('Loss vs Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

```python
test_generator=TimeseriesGenerator(data=scaled_test,
                                targets=scaled_test,
                                length=n_input,batch_size=32
                                )
```

Hyperparameteric tunning of RNN architecture

```python
from kerastuner.tuners import RandomSearch
from kerastuner.engine.hyperparameters import HyperParameters
from pickle import dump,load
import warnings
warnings.simplefilter("ignore", UserWarning)
from tensorflow.keras.models import save_model
from tensorflow.keras.models import model_from_json
```

```python



# Extracting X_train and y_train
X_train = generator[0][0]  # Accessing the first batch of input sequences
y_train = generator[0][1]  # Accessing the corresponding targets for the first batch

# If you have multiple batches, you can concatenate them
for i in range(1, len(generator)):
    X_batch, y_batch = generator[i]
    X_train = np.concatenate((X_train, X_batch), axis=0)
    y_train = np.concatenate((y_train, y_batch), axis=0)
```

```python

# Extracting X_train and y_train
X_test = test_generator[0][0]  # Accessing the first batch of input sequences
y_test = test_generator[0][1]  # Accessing the corresponding targets for the first batch

# If you have multiple batches, you can concatenate them
for i in range(1, len(test_generator)):
    X_batch, y_batch = test_generator[i]
    X_test = np.concatenate((X_test, X_batch), axis=0)
    y_test = np.concatenate((y_test, y_batch), axis=0)
```

```python
def build_model(hp):
    model = Sequential()
    model.add(SimpleRNN(hp.Int('input_unit',min_value=32,max_value=512,step=32),return_sequences=True, input_shape=(X_train.shape[1],X_train.shape[2])))
    for i in range(hp.Int('n_layers', 1, 4)):
        model.add(SimpleRNN(hp.Int(f'rnn_{i}_units',min_value=32,max_value=512,step=32),return_sequences=True))
    model.add(SimpleRNN(hp.Int('input_unit',min_value=32,max_value=512,step=32),return_sequences=False))
    model.add(Dense(2, activation='relu'))
    model.compile(loss='mean_squared_error', optimizer='adam', metrics=[keras.metrics.MeanAbsoluteError()])
    return model
```

```python
import keras_tuner 

tuner= RandomSearch(
        build_model,
        objective=keras_tuner.Objective("val_mean_absolute_error", direction="min"),
        max_trials=2,
        executions_per_trial=1
        )
```

```python
tuner.search(
        x=X_train,
        y=y_train,
        epochs=20,
        validation_data=(X_test,y_test),
        batch_size=16
    
)
```

```python
best_model = tuner.get_best_models(num_models=1)[0]
```
